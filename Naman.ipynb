```python
import pandas as pd         #Import Pandas
import numpy as np
d = pd.read_csv('/content/drive/MyDrive/Copy of Data.csv')      #Load Data

d_copy = d.copy()       #Copy Data

l=[3,4,6,9,10,11,14,15,16,17,18,19,20,23,24,25,26,30,31]
d_copy.drop(d_copy.columns[l], axis=1, inplace=True)        #Remove Columns

#Rename Columns
d_copy.rename(columns={'CMPLNT_NUM' : 'cmp_no', 'CMPLNT_FR_DT' : 'cmp_dt', 'CMPLNT_FR_TM' : 'cmp_tm', 'ADDR_PCT_CD' : 'police_station', 'KY_CD' : 'crime_code', 'OFNS_DESC' : 'crime_name', 'LAW_CAT_CD' : 'crime_level', 'BORO_NM': 'city_name', 'X_COORD_CD' : 'x_cord', 'Y_COORD_CD' : 'y cord', 'VIC_AGE_GROUP' :'vic_age_group', 'VIC_RACE':' vic_cast', 'VIC_SEX':'vic_sex'}, inplace=True)

columns_list = d_copy.columns.tolist()      #list columns 

d_copy = d_copy[~d_copy['city_name'].isin(['nan'])]
d_copy = d_copy[~d_copy['police_station'].isin(['nan'])]
d_copy = d_copy[d_copy['vic_age_group'].isin(['<18', '18-24','25-44','45-64','65+'])]       #Remove Unknown Values
d_copy = d_copy[~d_copy[' vic_cast'].isin(['UNKNOWN'])]
d_copy = d_copy[d_copy['vic_sex'].isin(['M','F'])]



a = np.ones(5158938)
d_copy['result'] = a

d_copy['cmp_tm_hr'] = d_copy['cmp_tm'].str[0:2]
d_copy['cmp_tm_min'] = d_copy['cmp_tm'].str[3:5]
d_copy['cmp_tm_sec'] = d_copy['cmp_tm'].str[6:8]
d_copy['cmp_dt_month'] = d_copy['cmp_dt'].str[0:2]
d_copy['cmp_dt_day'] = d_copy['cmp_dt'].str[3:5]
d_copy['cmp_dt_year'] = d_copy['cmp_dt'].str[6:10]

columns_list = d_copy.columns.tolist() 

d_copy['cmp_tm_hr'] = d_copy['cmp_tm_hr'].fillna(0).astype(int)
d_copy['cmp_tm_min'] = d_copy['cmp_tm_min'].fillna(0).astype(int)
d_copy['cmp_tm_sec'] = d_copy['cmp_tm_sec'].fillna(0).astype(int)
d_copy['cmp_dt_day'] = d_copy['cmp_dt_day'].fillna(0).astype(int)
d_copy['cmp_dt_month'] = d_copy['cmp_dt_month'].fillna(0).astype(int)
d_copy['cmp_dt_year'] = d_copy['cmp_dt_year'].fillna(0).astype(int)

d_copy['crime_name'], unique_values = pd.factorize(d_copy['crime_name'])
d_copy['crime_level'], unique_values = pd.factorize(d_copy['crime_level'])
d_copy['city_name'], unique_values = pd.factorize(d_copy['city_name'])
d_copy['vic_age_group'], unique_values = pd.factorize(d_copy['vic_age_group'])
d_copy[' vic_cast'], unique_values = pd.factorize(d_copy[' vic_cast'])
d_copy['vic_sex'], unique_values = pd.factorize(d_copy['vic_sex'])

new_order=['cmp_no','cmp_dt_day','cmp_dt_month', 'cmp_dt_year'  , 'cmp_tm_hr', 'cmp_tm_min', 'cmp_tm_sec',  'x_cord', 'y cord', 'Latitude', 'Longitude', 'Lat_Lon', 'vic_age_group', ' vic_cast',
       'vic_sex', 'city_name', 'police_station', 'crime_level','crime_code',  'crime_name',   'result', 'cmp_dt_Month', 'cmp_dt_Year', 'cmp_dt', 'cmp_tm']
d_copy = d_copy.reindex(columns=new_order)

d_copy.drop(columns=['cmp_dt_Month'], inplace=True)
d_copy.drop(columns=['cmp_dt_Year'], inplace=True)
d_copy.drop(columns=['cmp_dt'], inplace=True)
d_copy.drop(columns=['cmp_tm'], inplace=True)
d_copy.drop(columns=['Lat_Lon'], inplace=True)

cols = d_copy.columns.tolist()
for i in cols:

    d_copy = d_copy[~d_copy[i].isin([-1])]
    d_copy = d_copy[~d_copy[i].isin(['nan'])]

print(d_copy)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Input(input_shape=(2,)),
                         tf.keras.layers.Dense(64, activation='relu'),
                         tf.keras.layers.Dense(64, activation='relu'),
                         tf.keras.layers.Dense(2)])


model.compile(optimizer='adam',
              loss= tf.keras.losses.SparseCategoricalCrossentropy(), 
              metrics=['accuracy'])
```

```python
#y_train_encoded = tf.keras.utils.to_categorical(y_train)
model.fit(X_test, y_test, epochs=10)


predictions = model.predict(X_test)


y_test_encoded = tf.keras.utils.to_categorical(y_test)
loss, accuracy = model.evaluate(X_test, y_test_encoded)


print("Loss:", loss)
print("Accuracy:", accuracy)
```
